```
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt

np.random.seed(1)
torch.manual_seed(1)

# generate synthetic data
T = 20
L = 1000
N = 100

x = np.tile(np.array(range(L)), (N, 1)) + np.tile(np.random.randint(-4*T, 4*T, N).reshape(N, 1), (1, L))
data = np.sin(x / 1.0 / T)


class RNN(nn.Module):
  def __init__(self):
    super(RNN, self).__init__()
    self.hidden_size = 32
    self.rnn1 = nn.RNNCell(input_size=1, hidden_size=self.hidden_size)
    self.rnn2 = nn.RNNCell(input_size=self.hidden_size, hidden_size=self.hidden_size)
    self.fc = nn.Linear(in_features=self.hidden_size, out_features=1)

  def forward(self, x, device, future_step = 0):
    outputs = []
    h_t1 = torch.zeros(x.size(0), self.hidden_size).to(device)
    h_t2 = torch.zeros(x.size(0), self.hidden_size).to(device)

    for _, x_t in enumerate(x.chunk(x.size(1), dim=1)):
      h_t1 = self.rnn1(x_t, h_t1)
      h_t2 = self.rnn2(h_t1, h_t2)
      output = self.fc(h_t2)
      outputs += [output]
    for i in range(future_step):
      h_t1 = self.rnn1(output, h_t1)
      h_t2 = self.rnn2(h_t1, h_t2)
      output = self.fc(h_t2)
      outputs += [output]
    outputs = torch.stack(outputs, dim=1).squeeze(2)

    return outputs

class LSTM(nn.Module):
  def __init__(self):
    super(LSTM, self).__init__()
    self.hidden_size = 32
    self.lstm1 = nn.LSTMCell(input_size=1, hidden_size=self.hidden_size)
    self.lstm2 = nn.LSTMCell(input_size=self.hidden_size, hidden_size=self.hidden_size)
    self.fc = nn.Linear(in_features=self.hidden_size, out_features=1)

  def forward(self, x, device, future_step = 0):
    outputs = []

    h_t1 = torch.zeros(x.size(0), self.hidden_size).to(device)
    c_t1 = torch.zeros(x.size(0), self.hidden_size).to(device)
    h_t2 = torch.zeros(x.size(0), self.hidden_size).to(device)
    c_t2 = torch.zeros(x.size(0), self.hidden_size).to(device)

    for _, x_t in enumerate(x.chunk(x.size(1), dim=1)):
      h_t1, c_t1 = self.lstm1(x_t, (h_t1, c_t1))
      h_t2, c_t2 = self.lstm2(h_t1, (h_t2, c_t2))
      output = self.fc(h_t2)
      outputs += [output]
    for i in range(future_step):
      h_t1, c_t1 = self.lstm1(output, (h_t1, c_t1))
      h_t2, c_t2 = self.lstm2(h_t1, (h_t2, c_t2))
      output = self.fc(h_t2)
      outputs += [output]
    outputs = torch.stack(outputs, dim=1).squeeze(2)

    return outputs

def train(model, device, train_dataset, optimizer, epoch, batch_size):
    model.train()
    data, target = train_dataset
    data, target = torch.Tensor(data), torch.Tensor(target)
    for batch_idx in range(data.size(0) // batch_size + 1):
        data_i, target_i = data[batch_idx*batch_size:(batch_idx+1)*batch_size], target[batch_idx*batch_size:(batch_idx+1)*batch_size]
        data_i, target_i = data_i.to(device), target_i.to(device)
        def closure():
          optimizer.zero_grad()
          output = model(data_i, device)
          loss = F.mse_loss(output, target_i)
          loss.backward()
          print('Train Epoch: {} {}/{} \tLoss: {:.6f}'.format(
              epoch, batch_idx * batch_size + data_i.size(0), data.size(0), loss.item()))
          return loss
        optimizer.step(closure)
        
  
def test(model, device, test_dataset, batch_size):
    model.eval()
    test_loss = 0
    mae = 0
    data, target = test_dataset
    data, target = torch.Tensor(data), torch.Tensor(target)
    with torch.no_grad():
        for batch_idx in range(data.size(0) // batch_size + 1):
            data_i, target_i = data[batch_idx*batch_size:(batch_idx+1)*batch_size], target[batch_idx*batch_size:(batch_idx+1)*batch_size]
            data_i, target_i = data_i.to(device), target_i.to(device)
            output = model(data_i, device, future_step = 100)
            recon, predict = output[:, :-100], output[:, -100:]
            test_loss += F.mse_loss(recon, target_i).item()  # sum up batch loss
            mae += torch.abs(recon.cpu() - target_i.cpu()).mean()

    test_loss /= (data.size(0) // batch_size + 1)
    mae /= (data.size(0) // batch_size + 1)

    print('\nTest set: Average loss: {:.4f}, MAE: {:.4f}\n'.format(
        test_loss, mae))
    x, xf = np.array(range(1, L)), np.array(range(L, L+100))
    yr, yp1, yp2 = target_i[0].cpu().numpy(), recon[0].cpu().numpy(), predict[0].cpu().numpy()
    plt.plot(x, yr, 'ro', x, yp1, 'bo', xf, yp2, 'bo')
    plt.pause(1)
    plt.close()

def main():
    # Hyperparameter Settings
    batch_size = 100
    epochs = 100
    learning_rate = 0.8

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    train_dataset = (data[3:, :-1], data[3:, 1:])
    test_dataset = (data[:3, :-1], data[:3, 1:])

    model = RNN().to(device)
    optimizer = torch.optim.LBFGS(model.parameters(), lr=learning_rate, max_iter = 5)

    for epoch in range(1, epochs + 1):
        train(model, device, train_dataset, optimizer, epoch, batch_size)
        test(model, device, test_dataset, batch_size)

if __name__ == '__main__':
    main()
  
```
